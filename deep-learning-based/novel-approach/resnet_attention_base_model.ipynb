```python
# DRASTICALLY IMPROVED MEDICAL IMAGE FUSION
# Fixes: Model architecture, loss function, training strategy

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.models as models
import torchvision.transforms as transforms
import torchvision.transforms.functional as TF
from torch.cuda import amp
from tqdm import tqdm
import cv2
import numpy as np
from PIL import Image
import os
import random
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Fix multiprocessing
import torch.multiprocessing as mp
try:
    mp.set_start_method('spawn', force=True)
    print("Set multiprocessing start method to 'spawn'")
except RuntimeError:
    pass

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# ================================
# IMPROVED CONFIGURATION
# ================================

config = {
    'batch_size': 8,         # Increased for stability
    'lr': 2e-4,             # Higher learning rate
    'num_epochs': 50,
    'device': device,
    'save_path': 'fusion_model_fixed.pth',
    'image_size': (256, 256),
    'num_workers': 0,       # Avoid multiprocessing issues
    
    # COMPLETELY REVISED LOSS WEIGHTS
    'w_content': 10.0,      # NEW: Content preservation (most important)
    'w_structure': 5.0,     # NEW: Structural similarity 
    'w_detail': 3.0,        # NEW: Detail preservation
    'w_consistency': 2.0,   # NEW: Cross-modal consistency
    'w_gradient': 1.0,      # Reduced gradient weight
}

print("Improved Configuration:")
for key, value in config.items():
    print(f'  {key}: {value}')

# ================================
# IMPROVED DATASET WITH BETTER PREPROCESSING
# ================================

class ImprovedLLVIPDataset(Dataset):
    def __init__(self, root_dir, train=True):
        self.train = train
        mode = 'train' if train else 'test'
        self.vis_path = os.path.join(root_dir, 'visible', mode)
        self.ir_path = os.path.join(root_dir, 'infrared', mode)
        
        self.vis_images = sorted([f for f in os.listdir(self.vis_path) 
                                 if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif'))])
        self.ir_images = sorted([f for f in os.listdir(self.ir_path) 
                                 if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif'))])
        
        self.pairs = [f for f in self.vis_images if f in self.ir_images]
        print(f"Found {len(self.pairs)} image pairs in {mode} set")

        # IMPROVED TRANSFORMS - NO NORMALIZATION IN DATASET
        if train:
            self.transform = transforms.Compose([
                transforms.Resize(config['image_size']),
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1),
                transforms.ToTensor(),  # [0,1]
            ])
        else:
            self.transform = transforms.Compose([
                transforms.Resize(config['image_size']),
                transforms.ToTensor(),  # [0,1]
            ])

    def __len__(self):
        return len(self.pairs)
    
    def __getitem__(self, idx):
        img_name = self.pairs[idx]
        
        vis_img = Image.open(os.path.join(self.vis_path, img_name)).convert('RGB')
        ir_img = Image.open(os.path.join(self.ir_path, img_name)).convert('L')
        
        # Apply synchronized transforms
        if self.train:
            # Random transformations with same seed
            seed = random.randint(0, 2**32)
            random.seed(seed)
            torch.manual_seed(seed)
            vis_tensor = self.transform(vis_img)
            
            random.seed(seed)
            torch.manual_seed(seed)
            ir_tensor = self.transform(ir_img)
        else:
            vis_tensor = self.transform(vis_img)
            ir_tensor = self.transform(ir_img)
        
        return vis_tensor, ir_tensor, img_name

# ================================
# COMPLETELY REDESIGNED MODEL ARCHITECTURE
# ================================

class SpatialAttention(nn.Module):
    """Spatial attention to focus on important regions"""
    def __init__(self, in_channels):
        super().__init__()
        self.conv = nn.Conv2d(in_channels, 1, kernel_size=7, padding=3)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        attention = self.conv(x)
        attention = self.sigmoid(attention)
        return x * attention

class ChannelAttention(nn.Module):
    """Channel attention for feature selection"""
    def __init__(self, in_channels, reduction=16):
        super().__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(in_channels, in_channels // reduction, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(in_channels // reduction, in_channels, bias=False),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        b, c, _, _ = x.size()
        y = self.avg_pool(x).view(b, c)
        y = self.fc(y).view(b, c, 1, 1)
        return x * y.expand_as(x)

class DualAttentionFusion(nn.Module):
    """Improved fusion block with dual attention"""
    def __init__(self, channels):
        super().__init__()
        self.channel_attention = ChannelAttention(channels)
        self.spatial_attention = SpatialAttention(channels)
        
        self.fusion_conv = nn.Sequential(
            nn.Conv2d(channels * 2, channels, 3, padding=1),
            nn.BatchNorm2d(channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(channels, channels, 3, padding=1),
            nn.BatchNorm2d(channels),
            nn.ReLU(inplace=True)
        )
        
    def forward(self, vis_feat, ir_feat):
        # Apply attention to each modality
        vis_att = self.spatial_attention(self.channel_attention(vis_feat))
        ir_att = self.spatial_attention(self.channel_attention(ir_feat))
        
        # Concatenate and fuse
        combined = torch.cat([vis_att, ir_att], dim=1)
        fused = self.fusion_conv(combined)
        
        return fused

class ImprovedFusionModel(nn.Module):
    """Completely redesigned fusion architecture"""
    def __init__(self):
        super().__init__()
        
        # SHARED ENCODER APPROACH (better than separate encoders)
        backbone = models.resnet34(pretrained=True)
        
        # RGB encoder (keep pretrained weights)
        self.vis_stem = nn.Sequential(
            backbone.conv1,      # 64 channels
            backbone.bn1,
            backbone.relu,
            backbone.maxpool
        )
        self.vis_layer1 = backbone.layer1  # 64
        self.vis_layer2 = backbone.layer2  # 128  
        self.vis_layer3 = backbone.layer3  # 256
        self.vis_layer4 = backbone.layer4  # 512
        
        # IR encoder (adapt first layer for single channel)
        ir_backbone = models.resnet34(pretrained=True)
        # Adapt conv1 for single channel IR input
        old_conv1 = ir_backbone.conv1
        ir_backbone.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
        # Initialize with average of RGB weights
        with torch.no_grad():
            ir_backbone.conv1.weight = nn.Parameter(old_conv1.weight.mean(dim=1, keepdim=True))
            
        self.ir_stem = nn.Sequential(
            ir_backbone.conv1,
            ir_backbone.bn1, 
            ir_backbone.relu,
            ir_backbone.maxpool
        )
        self.ir_layer1 = ir_backbone.layer1
        self.ir_layer2 = ir_backbone.layer2
        self.ir_layer3 = ir_backbone.layer3
        self.ir_layer4 = ir_backbone.layer4
        
        # Fusion blocks at each level
        self.fusion1 = DualAttentionFusion(64)
        self.fusion2 = DualAttentionFusion(128)
        self.fusion3 = DualAttentionFusion(256)
        self.fusion4 = DualAttentionFusion(512)
        
        # IMPROVED DECODER with skip connections
        self.decoder4 = nn.Sequential(
            nn.ConvTranspose2d(512, 256, 4, stride=2, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True)
        )
        
        self.decoder3 = nn.Sequential(
            nn.Conv2d(256 + 256, 256, 3, padding=1),  # Skip connection
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.ConvTranspose2d(256, 128, 4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True)
        )
        
        self.decoder2 = nn.Sequential(
            nn.Conv2d(128 + 128, 128, 3, padding=1),  # Skip connection
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.ConvTranspose2d(128, 64, 4, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True)
        )
        
        self.decoder1 = nn.Sequential(
            nn.Conv2d(64 + 64, 64, 3, padding=1),  # Skip connection
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.ConvTranspose2d(64, 32, 4, stride=2, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True)
        )
        
        # Final output layer - FIXED: Added upsampling to match input size
        self.final = nn.Sequential(
            nn.Upsample(scale_factor=2, mode='bilinear', align_corners=False),  # Added to match input size
            nn.Conv2d(32, 16, 3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(inplace=True),
            nn.Conv2d(16, 3, 1),
            nn.Sigmoid()  # Output in [0,1]
        )
        
    def forward(self, vis, ir):
        # Encode visible
        v0 = self.vis_stem(vis)      # 64, H/4, W/4
        v1 = self.vis_layer1(v0)     # 64, H/4, W/4  
        v2 = self.vis_layer2(v1)     # 128, H/8, W/8
        v3 = self.vis_layer3(v2)     # 256, H/16, W/16
        v4 = self.vis_layer4(v3)     # 512, H/32, W/32
        
        # Encode infrared  
        i0 = self.ir_stem(ir)        # 64, H/4, W/4
        i1 = self.ir_layer1(i0)      # 64, H/4, W/4
        i2 = self.ir_layer2(i1)      # 128, H/8, W/8  
        i3 = self.ir_layer3(i2)      # 256, H/16, W/16
        i4 = self.ir_layer4(i3)      # 512, H/32, W/32
        
        # Fuse at each level
        f1 = self.fusion1(v1, i1)    # 64
        f2 = self.fusion2(v2, i2)    # 128
        f3 = self.fusion3(v3, i3)    # 256  
        f4 = self.fusion4(v4, i4)    # 512
        
        # Decode with skip connections
        d4 = self.decoder4(f4)                           # 256, H/16, W/16
        d3 = self.decoder3(torch.cat([d4, f3], dim=1))   # 128, H/8, W/8
        d2 = self.decoder2(torch.cat([d3, f2], dim=1))   # 64, H/4, W/4
        d1 = self.decoder1(torch.cat([d2, f1], dim=1))   # 32, H/2, W/2
        
        output = self.final(d1)                           # 3, H, W
        
        return output

# ================================
# COMPLETELY REDESIGNED LOSS FUNCTION
# ================================

class ImprovedFusionLoss(nn.Module):
    """Much better loss function focused on meaningful fusion"""
    def __init__(self):
        super().__init__()
        self.l1_loss = nn.L1Loss()
        self.mse_loss = nn.MSELoss()
        
        # Sobel operators for gradient computation
        sobel_x = torch.tensor([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], dtype=torch.float32)
        sobel_y = torch.tensor([[-1, -2, -1], [0, 0, 0], [1, 2, 1]], dtype=torch.float32)
        
        self.register_buffer('sobel_x', sobel_x.view(1, 1, 3, 3))
        self.register_buffer('sobel_y', sobel_y.view(1, 1, 3, 3))
        
    def rgb_to_gray(self, rgb):
        """Convert RGB to grayscale"""
        return 0.299 * rgb[:, 0:1, :, :] + 0.587 * rgb[:, 1:2, :, :] + 0.114 * rgb[:, 2:3, :, :]
        
    def compute_gradient(self, img):
        """Compute gradient magnitude"""
        if img.size(1) == 3:
            img = self.rgb_to_gray(img)
        # Ensure Sobel filters are on the same device and dtype as input
        sobel_x = self.sobel_x.to(img.dtype).to(img.device)
        sobel_y = self.sobel_y.to(img.dtype).to(img.device)
        grad_x = F.conv2d(img, sobel_x, padding=1)
        grad_y = F.conv2d(img, sobel_y, padding=1)
        return torch.sqrt(grad_x**2 + grad_y**2 + 1e-8)
        
    def ssim_loss(self, x, y, window_size=11):
        """SSIM loss between two images"""
        if x.size(1) == 3:
            x = self.rgb_to_gray(x)
        if y.size(1) == 3:
            y = self.rgb_to_gray(y)
            
        mu_x = F.avg_pool2d(x, window_size, stride=1, padding=window_size//2)
        mu_y = F.avg_pool2d(y, window_size, stride=1, padding=window_size//2)
        
        sigma_x = F.avg_pool2d(x**2, window_size, stride=1, padding=window_size//2) - mu_x**2
        sigma_y = F.avg_pool2d(y**2, window_size, stride=1, padding=window_size//2) - mu_y**2
        sigma_xy = F.avg_pool2d(x*y, window_size, stride=1, padding=window_size//2) - mu_x*mu_y
        
        C1, C2 = 0.01**2, 0.03**2
        ssim_map = ((2*mu_x*mu_y + C1)*(2*sigma_xy + C2)) / ((mu_x**2 + mu_y**2 + C1)*(sigma_x + sigma_y + C2))
        
        return 1 - ssim_map.mean()
        
    def forward(self, fused, vis, ir):
        """
        Compute comprehensive fusion loss
        fused: (B, 3, H, W) - model output
        vis: (B, 3, H, W) - visible input  
        ir: (B, 1, H, W) - infrared input
        """
        
        # Convert IR to 3-channel for consistency
        ir_3ch = ir.repeat(1, 3, 1, 1)
        
        # FIXED: Ensure all tensors have the same spatial dimensions
        if fused.size(2) != vis.size(2) or fused.size(3) != vis.size(3):
            fused = F.interpolate(fused, size=vis.shape[2:], mode='bilinear', align_corners=False)
        
        # 1. CONTENT PRESERVATION - fused should contain info from both inputs
        vis_gray = self.rgb_to_gray(vis)
        ir_gray = ir
        fused_gray = self.rgb_to_gray(fused)
        
        # Fused image should be similar to the brighter of vis/ir at each pixel
        target = torch.max(vis_gray, ir_gray)
        content_loss = self.l1_loss(fused_gray, target)
        
        # 2. STRUCTURE PRESERVATION - maintain structural similarity
        structure_loss_vis = self.ssim_loss(fused, vis)
        structure_loss_ir = self.ssim_loss(fused_gray, ir_gray)
        structure_loss = 0.5 * (structure_loss_vis + structure_loss_ir)
        
        # 3. DETAIL PRESERVATION - preserve gradients/edges from both inputs
        grad_fused = self.compute_gradient(fused)
        grad_vis = self.compute_gradient(vis)
        grad_ir = self.compute_gradient(ir_gray)
        
        # Fused gradients should capture max gradients from either input
        grad_target = torch.max(grad_vis, grad_ir)
        detail_loss = self.l1_loss(grad_fused, grad_target)
        
        # 4. CONSISTENCY - fused image should not deviate too much from inputs
        consistency_loss_vis = self.mse_loss(fused, vis)
        consistency_loss_ir = self.mse_loss(fused, ir_3ch)
        consistency_loss = 0.5 * (consistency_loss_vis + consistency_loss_ir)
        
        # 5. GRADIENT PRESERVATION - specific gradient matching
        gradient_loss = 0.5 * (self.l1_loss(grad_fused, grad_vis) + self.l1_loss(grad_fused, grad_ir))
        
        # Combine losses with weights
        total_loss = (
            config['w_content'] * content_loss +
            config['w_structure'] * structure_loss +
            config['w_detail'] * detail_loss +
            config['w_consistency'] * consistency_loss +
            config['w_gradient'] * gradient_loss
        )
        
        return total_loss, {
            'content': content_loss.item(),
            'structure': structure_loss.item(), 
            'detail': detail_loss.item(),
            'consistency': consistency_loss.item(),
            'gradient': gradient_loss.item(),
            'total': total_loss.item()
        }

# ================================
# IMPROVED TRAINING FUNCTION
# ================================

def train_improved_model():
    print('Initializing improved training...')
    
    # Create improved model
    model = ImprovedFusionModel().to(config['device'])
    print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
    
    # Create datasets
    full_dataset = ImprovedLLVIPDataset('LLVIP', train=True)
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

    train_dataloader = DataLoader(
        train_dataset, 
        batch_size=config['batch_size'], 
        shuffle=True,
        num_workers=config['num_workers'], 
        pin_memory=True
    )
    
    val_dataloader = DataLoader(
        val_dataset,
        batch_size=config['batch_size'], 
        shuffle=False,
        num_workers=config['num_workers'], 
        pin_memory=True
    )

    # Optimizer with different learning rates for encoder vs decoder
    encoder_params = []
    decoder_params = []
    
    for name, param in model.named_parameters():
        if 'vis_' in name or 'ir_' in name:
            encoder_params.append(param)
        else:
            decoder_params.append(param)
    
    optimizer = optim.AdamW([
        {'params': encoder_params, 'lr': config['lr'] * 0.1},  # Lower LR for pretrained
        {'params': decoder_params, 'lr': config['lr']}         # Higher LR for new parts
    ], weight_decay=1e-4)
    
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=config['num_epochs'])
    scaler = amp.GradScaler()
    criterion = ImprovedFusionLoss()

    print(f'Training set: {len(train_dataset)} samples')
    print(f'Validation set: {len(val_dataset)} samples')
    print('Starting improved training loop...')

    best_val_loss = float('inf')
    training_history = []
    patience = 15
    patience_counter = 0

    for epoch in range(config['num_epochs']):
        # Training phase
        model.train()
        train_loss = 0.0
        train_components = {
            'content': 0.0, 'structure': 0.0, 'detail': 0.0, 
            'consistency': 0.0, 'gradient': 0.0
        }
        
        pbar = tqdm(train_dataloader, desc=f"Epoch {epoch+1}/{config['num_epochs']}")
        for vis_batch, ir_batch, _ in pbar:
            vis_batch = vis_batch.to(config['device'])  # (B, 3, H, W)
            ir_batch = ir_batch.to(config['device'])    # (B, 1, H, W)

            optimizer.zero_grad()
            
            with amp.autocast():
                fused_output = model(vis_batch, ir_batch)
                loss, loss_components = criterion(fused_output, vis_batch, ir_batch)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            train_loss += loss.item()
            for key in train_components:
                train_components[key] += loss_components[key]
                
            # Update progress bar
            pbar.set_postfix({
                'loss': f'{loss.item():.4f}',
                'content': f'{loss_components["content"]:.3f}',
                'structure': f'{loss_components["structure"]:.3f}'
            })

        # Average training losses
        avg_train_loss = train_loss / len(train_dataloader)
        for key in train_components:
            train_components[key] /= len(train_dataloader)

        # Validation phase
        model.eval()
        val_loss = 0.0
        val_components = {
            'content': 0.0, 'structure': 0.0, 'detail': 0.0,
            'consistency': 0.0, 'gradient': 0.0
        }
        
        with torch.no_grad():
            for vis_batch, ir_batch, _ in val_dataloader:
                vis_batch = vis_batch.to(config['device'])
                ir_batch = ir_batch.to(config['device'])
                
                with amp.autocast():
                    fused_output = model(vis_batch, ir_batch)
                    loss, loss_components = criterion(fused_output, vis_batch, ir_batch)
                
                val_loss += loss.item()
                for key in val_components:
                    val_components[key] += loss_components[key]

        avg_val_loss = val_loss / len(val_dataloader)
        for key in val_components:
            val_components[key] /= len(val_dataloader)
            
        scheduler.step()

        # Logging
        print(f"Epoch {epoch+1}:")
        print(f"  Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")
        print(f"  Content: {train_components['content']:.3f} | Structure: {train_components['structure']:.3f}")
        print(f"  Detail: {train_components['detail']:.3f} | LR: {optimizer.param_groups[0]['lr']:.6f}")

        training_history.append({
            'epoch': epoch + 1,
            'train_loss': avg_train_loss,
            'val_loss': avg_val_loss,
            **{f'train_{k}': train_components[k] for k in train_components},
            **{f'val_{k}': val_components[k] for k in val_components}
        })

        # Save best model
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save({
                'model_state_dict': model.state_dict(),
                'config': config,
                'epoch': epoch + 1,
                'best_val_loss': best_val_loss,
                'training_history': training_history
            }, config['save_path'])
            print(f"  ✓ New best model saved! (val_loss: {best_val_loss:.4f})")
            patience_counter = 0
        else:
            patience_counter += 1
            
        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch+1}")
            break
            
        print("-" * 60)

    print("\n" + "="*60)
    print("TRAINING COMPLETED!")
    print("="*60)
    print(f"Best validation loss: {best_val_loss:.4f}")
    print(f"Model saved to: {config['save_path']}")

    return model, training_history

# ================================
# VISUALIZATION AND TESTING
# ================================

def visualize_results(model, dataset, num_samples=4):
    """Visualize fusion results"""
    model.eval()
    
    # Create test dataloader
    test_loader = DataLoader(dataset, batch_size=1, shuffle=True, num_workers=0)
    
    fig, axes = plt.subplots(num_samples, 4, figsize=(16, 4*num_samples))
    if num_samples == 1:
        axes = axes.reshape(1, -1)
    
    with torch.no_grad():
        for idx, (vis, ir, name) in enumerate(test_loader):
            if idx >= num_samples:
                break
                
            vis = vis.to(config['device'])
            ir = ir.to(config['device'])
            
            # Generate fusion
            fused = model(vis, ir)
            
            # Convert to numpy for visualization
            vis_np = vis[0].cpu().permute(1, 2, 0).numpy()
            ir_np = ir[0, 0].cpu().numpy()
            fused_np = fused[0].cpu().permute(1, 2, 0).numpy()
            
            # Create comparison image (simple blend for reference)
            vis_gray = np.mean(vis_np, axis=2)
            comparison = np.stack([vis_gray, ir_np, vis_gray], axis=2)
            
            # Plot
            axes[idx, 0].imshow(vis_np)
            axes[idx, 0].set_title(f'Visible {idx+1}')
            axes[idx, 0].axis('off')
            
            axes[idx, 1].imshow(ir_np, cmap='gray')
            axes[idx, 1].set_title(f'Infrared {idx+1}')
            axes[idx, 1].axis('off')
            
            axes[idx, 2].imshow(fused_np)
            axes[idx, 2].set_title(f'Fused {idx+1}')
            axes[idx, 2].axis('off')
            
            axes[idx, 3].imshow(comparison)
            axes[idx, 3].set_title(f'Reference {idx+1}')
            axes[idx, 3].axis('off')
    
    plt.tight_layout()
    plt.savefig('improved_fusion_results.png', dpi=150, bbox_inches='tight')
    plt.show()

# ================================
# MAIN EXECUTION SCRIPT
# ================================

def main():
    """Main execution function"""
    print("="*60)
    print("IMPROVED MEDICAL IMAGE FUSION TRAINING")
    print("="*60)
    
    # Check if dataset exists
    if not os.path.exists('LLVIP'):
        print("ERROR: LLVIP dataset not found!")
        print("Please download and extract the LLVIP dataset to the current directory.")
        return
    
    # Train the improved model
    try:
        model, history = train_improved_model()
        
        # Create test dataset for evaluation
        test_dataset = ImprovedLLVIPDataset('LLVIP', train=False)
        
        # Visualize some results
        print("\nGenerating result visualizations...")
        visualize_results(model, test_dataset, num_samples=4)
        
        print("\n" + "="*60)
        print("TRAINING AND EVALUATION COMPLETED SUCCESSFULLY!")
        print("="*60)
        print(f"✓ Model saved to: {config['save_path']}")
        print("✓ Result visualizations saved to: improved_fusion_results.png")
        print("✓ Ready for deployment!")
        
        return model
        
    except Exception as e:
        print(f"\nERROR during training: {e}")
        import traceback
        traceback.print_exc()
        return None

if __name__ == "__main__":
    # Run the complete training and evaluation pipeline
    model = main()
    
    if model is not None:
        print("\n🎉 All tasks completed successfully!")
        print("\n📋 KEY IMPROVEMENTS MADE:")
        print("✓ Fixed model architecture with proper skip connections")
        print("✓ Redesigned loss function focusing on meaningful fusion")
        print("✓ Better training strategy with differential learning rates")
        print("✓ Fixed dimension mismatch in loss function")
        print("✓ Ready for meaningful medical image fusion!")
    
    print("\n🚀 The model should now produce meaningful fusion results instead of colorful noise!")
```
